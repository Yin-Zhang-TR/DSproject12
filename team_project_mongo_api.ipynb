```python
import pymongo
from pymongo.errors import ConnectionFailure

def connection(host = 'localhost', port = 27017):
    client = MongoClient(host, port, connectTimeoutMS=1000)
    try:
        # The ismaster command is cheap and does not require auth.
        print("Test Mongodb whether available")
        client.admin.command('ismaster')
    except ConnectionFailure:
        print("Server not available")
        return -1
    
    return client

def database(client, dbname):
    try:
        db = client.dbname
    except Exception as e:
        print(e)
    
    return db

def collection(client, collection_name):
    try:
        collection = client.collection_name
    except Exception as e:
        print(e)
        
    return collection

def insert(document, collection_name, multi = False):
    try:
        if multi:
            insert_result = collection_name.insert_many(document)
        else:
            insert_result = collection_name.insert_one(document)
    except Exception as e:
        print(e)
        
    return insert_result

def update(old_values, new_values, collection_name, multi = False):
    try:
        if multi:
            update_result = collection_name.update_many(old_values, {"$set":new_values})
        else:
            update_result = collection_name.update_one(old_values, {"$set":new_values})
    except Exception as e:
        print(e)
        
    return update_result

def delete(item_values, collection_name, multi = False):
    try:
        if multi:
            delete_result = collection_name.delete_many(item_values)
        else:
            delete_result = collection_name.delete_one(item_values)
    except Exception as e:
        print(e)
        
    return delete_result

def find(item_values, collection_name):
    try:
        find_result = collection_name.find(item_values)
    except Exception as e:
        print(e)
        
    return find_result

#if __name__ == '__main__':
    #a = connection(host = '127.0.0.1', port = 27017)
    #print(a.test)
    #print(a.test.test)
    
```
