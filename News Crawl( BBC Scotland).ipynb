```python
def news_content(news_url):
    c=[]
    c1=[]
    res = requests.get(news_url, headers = headers)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text, 'lxml')

    for a in soup.select('.story-body__inner'):
        if (len(a.select('p'))>5):
            for i in range(5):
                content=a.select('p')[i].text
                print(content)
                c.append(content)
            return(c)
        else:
            for j in range(len(a.select('p'))):
                content=a.select('p')[j].text
                print(content)
                c1.append(content)
            return(c1)
```

# Time

```python
from datetime import datetime
def time_transfer(time):
    ts = int(time)

    # if you encounter a "year is out of range" error the timestamp
    # may be in milliseconds, try `ts /= 1000` in that case
    return(datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
article_url=[]
article_title=[]
article_time=[]
article_content=[]
for i in range(len(soup.select('.title-link'))):
    a=soup.select('.title-link')[i]
    url=a['href']
    title=a.text
    if 'https://' in url:
        print(title)
        print(url)
        article_url.append(url)
        article_time.append(time(url))
        article_content.append(news_content(url))
    else:
        url='https://www.bbc.co.uk'+url
        print(title)
        print(url)
        article_url.append(url)
        article_time.append(time(url))
        article_content.append(news_content(url))
    article_title.append(title)
    
#len(article_url)
```
